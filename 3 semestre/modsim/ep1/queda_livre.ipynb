```python
def Analitica(x0, t0, v0, dt):
  x = x0;
  t = t0;
  v = v0;
  g = -9.8;
  vxa = []
  vva = []
  while(x >= 0):
    x = x0 + v0*t + (1/2)*t*t*g
    v = v0 + g*t
    vxa.append(x)
    vva.append(v)
    t += dt

  return vxa, vva

vxa, vva = Analitica(3, 0, 0, 0.2)

plt.figure(0)
plt.plot(vxa, linestyle='', marker='o') 
plt.title('Posição (em metros)')
plt.show()

plt.figure(1)
plt.plot(vva, linestyle='', marker='o') 
plt.title('Velocidade (em metros/s)')
plt.show()
```

```python
def Euler(x0, t0, v0, tf, dt, acelerometro):
  x = x0
  t = t0*100
  t = int(t)
  tf *= 100
  tf = int(tf)
  v = v0
  vxe = []
  vve = []
  vae = []
  i = 0
  while(t < tf - dt):
    vxe.append(x)
    vve.append(v)
    dt *= 100
    dt = int(dt)
    t += dt
    aceleracao = acelerometro[t:(t + dt),4]
    a = aceleracao[0]*(9.8)-(9.8)
    vae.append(a)
    dt = dt/100
    v = v + a*dt
    x = x + v*dt
  
  return vxe, vve, vae

vxe, vve, vae = Euler(3.2, 51.75, 0, 53.75, 0.01, acelerometro)

plt.figure(0)
plt.plot(vxe) 
plt.title('Posição')
plt.show()

plt.figure(1)
plt.plot(vve) 
plt.title('Velocidade')
plt.show()

plt.figure(2)
plt.plot(vae) 
plt.title('Aceleração')
plt.show()


     
```
